#### 2). [Testing Encoding schemes] Compute weights for words, sentences, paragraphs using Google USE or TF IDF

##### - Google Universal Sentence encoder

```python
import tensorflow as tf
import tensorflow_hub as hub
embed=hub.load('./use/')
```

```python
#word=["what is your age","how old are you"]
#word=['A group of kids are playing in a yard and an old man is standing in the background',#      'The young boys are playing outdoors and the man is smiling nearby']

word=["Hello good morning"]

embeddings = embed(word)
#print(embeddings)
print(embeddings)
```

##### - TF-IDF(Term Frequency Inverse Document Frequency) encoding Scheme

```python
from sklearn.feature_extraction.text import TfidfVectorizer

Tfi = TfidfVectorizer()
tfidEncoding = Tfi.fit_transform(word)
#transT=trans.transform(['the'])
#print(tfidEncoding)
print(tfidEncoding.shape)
```

### Testing cosine similarity measure performance for Google USE and TFIDF encodings

```python
from sklearn.metrics.pairwise import cosine_similarity
print(word)
x=embeddings[0]
y=embeddings[1]
print(f"Google USE Score: {cosine_similarity([x],[y])}")
m=tfidEncoding[0]
n=tfidEncoding[1]
print(f"TF-IDF Score: {cosine_similarity(m,n)}")
```
